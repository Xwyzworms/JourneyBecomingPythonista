```python
v = np.array([30, -20 ])
lambdas = 10
vmodulated = v * lambdas

plt.plot([0,v[0]] , [0,v[1]])
plt.plot([0,vmodulated[0]] + v[0] , [0,vmodulated[1]]+  v[1],"k--")
max_modulated = max( abs(vmodulated * 1.2))
plt.axis((-max_modulated,max_modulated, -max_modulated,max_modulated) )
plt.grid()
```

```python
v1 = np.array([1,2,3,4,5])
v2 = np.array([2,3,4,5,6])
print(v1)
d1 = sum(np.multiply(v1.T,v2))
d2 = np.dot(v1.T,v2)
d3 = np.matmul(v1.T,v2)
d4 = v1.T @ v2
print(d1,d2,d3,d4)
```

```python
# Dot product is DIstributive
vecA = np.random.randn(5)
vecB = np.random.randn(5)
vecC = np.random.randn(5)

FirstCal = np.dot(vecA  , (vecB +  vecC))
SecondCal = np.dot(vecA , vecB) + np.dot(vecA, vecC)

print(FirstCal, SecondCal)
```

```python
# Dot product is not Associative
vecA =  np.random.randn(5)
vecB = np.random.randn(5)
vecC = np.random.randn(5)

FirstCal = np.dot(vecC, np.dot(vecA,vecB) )
SecondCal = np.dot(np.dot(vecB,vecC) , vecA)
print(FirstCal,SecondCal)
```

```python
# Dot product using Matrix
np.random.seed(5)
A = np.random.randn(4,6)
B = np.random.randn(4,6)
print(A.shape)
print(B.shape)
ans = np.zeros(shape= (6) )
for i in range(6):
    ans[i] = np.dot(A[:,i].T , B[:,i])
print(ans)
```

```python
# Dot product Is commutative
vecA = np.random.rand(5)
vecB = np.random.rand(5)

print( np.dot(vecA.T,vecB)) 
print( np.dot(vecB.T, vecA))

```
